```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import warnings

warnings.filterwarnings(
    "ignore", category=DeprecationWarning
)  # to avoid deprecation warnings
```

```python
# Download dataset
path = "https://fullstackds-projects-bucket.s3.eu-west-3.amazonaws.com/data/glassdoor_reviews.csv"
df = pd.read_csv(path)
df.head()
```

```python
# Filter out only 'Pizza-Hut' as firm
pizza_df = df[df['firm'] == "Pizza-Hut"].copy()
```

```python
# Keep only columns necessary for model prediction
pizza_demo = pizza_df[['headline', 'pros', 'cons']].dropna(subset=['pros', 'cons'])
```

```python
pizza_demo.to_csv("s3://fullstackds-projects-bucket/data/new_company_reviews_full.csv", index=False)
```

```python
# Sample of 500 entries
pizza_demo_sample = pizza_demo.sample(n=500, random_state=42)
```

```python
pizza_demo_sample.to_csv("s3://fullstackds-projects-bucket/data/new_company_reviews_sample.csv", index=False)
```
